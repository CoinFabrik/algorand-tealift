```python
from IPython.core.display import SVG
from subprocess import run

def tealift_draw(filepath: str):
    graph = run(["ts-node", "src/draw.ts", filepath], capture_output=True).stdout
    svg = run(["dot", "-Tsvg"], capture_output=True, input=graph).stdout
    return SVG(data=svg)

def tealift_print(filepath: str):
    return run(["ts-node", "src/print.ts", filepath], capture_output=True).stdout.decode()
```

```python
tealift_draw("tests/polynomial.teal")
```
